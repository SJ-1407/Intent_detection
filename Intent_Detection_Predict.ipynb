```python
import torch
import pandas as pd
```

```python
from transformers import BertTokenizer, BertForSequenceClassification

model_path_1 = "SJ-14/Intent_Model"
model_1 = BertForSequenceClassification.from_pretrained(model_path_1)
tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_1.eval()
```

```python
input_sentence = "What is the Pillow?"
```

```python
dataset = pd.read_csv('/content/sofmattress_train.csv')
label_map = {label: idx for idx, label in enumerate(dataset['label'].unique())}
```

```python
reverse_label_map = {idx: label for label, idx in label_map.items()}

```

```python
def predict_intent(input_sentence):
  encoding = tokenizer(
    input_sentence,
    padding='max_length',
    truncation=True,
    max_length=128,
    return_tensors="pt"
  )
  with torch.no_grad():
    outputs = model_1(
        input_ids=encoding['input_ids'],
        attention_mask=encoding['attention_mask']
    )
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=-1).item()

    return predicted_class
```

```python
predicted_class=predict_intent(input_sentence)
```

```python
predicted_intent = reverse_label_map[predicted_class]

print(f"Input Sentence: {input_sentence}")
print(f"Predicted Intent: {predicted_intent}")
```
